# Ê•µÊÄßËæûÊõ∏„Å®„Åó„Å¶Ôºå„ÄåÂçòË™ûÊÑüÊÉÖÊ•µÊÄßÂÄ§ÂØæÂøúË°®„Äç„ÇíÁî®„ÅÑ„Çã

```python
pndic = pd.read_csv(r"http://www.lr.pi.titech.ac.jp/~takamura/pubs/pn_ja.dic",
                    encoding="shift-jis",
                    names=['word_type_score'])

import numpy as np

# Ë™û„Å®ÊÑüÊÉÖÂÄ§„ÇíÊäΩÂá∫
pndic["split"] = pndic["word_type_score"].str.split(":")
pndic["word"] = pndic["split"].str.get(0)
pndic["score"] = pndic["split"].str.get(3)

# dictÂûã„Å´Â§âÊèõ
keys = pndic['word'].tolist()
values = pndic['score'].tolist()
dic = dict(zip(keys, values))

# print(dic)
```

```python
df_descrip = df_topic_nanacomi_new["denoising_description_new"]
```

```python
mean_list_main = []
for i in range(len(df_descrip)):
  text = df_descrip[i]
  try:
    lines = text.split("„ÄÇ")
  except AttributeError:
    lines = text
  # ÂΩ¢ÊÖãÁ¥†Ëß£Êûê„Å´Âü∫„Å•„ÅÑ„Å¶ÂçòË™û„ÇíÊäΩÂá∫
  # „Åæ„Åö„ÅØÔºåÊ•µÊÄßÂàÜÊûê„ÅÆ„Åü„ÇÅ„Å´ÂΩ¢ÊÖãÁ¥†Ëß£Êûê„ÇíË°å„ÅÜ
  word_list = []
  try:
    for l in lines:
        temp = []
        for v in mecab.parse(l).splitlines():
            if len(v.split()) >= 4:
                if v.split()[4][:2] in ['ÂêçË©û','ÂΩ¢ÂÆπË©û','ÂãïË©û','ÂâØË©û']:
                    temp.append(v.split()[3])
        word_list.append(temp)

    # Á©∫„ÅÆË¶ÅÁ¥†„ÇíÂâäÈô§
    word_list = [x for x in word_list if x != []]

    #print(word_list)
    result = []
    # ÊñáÂçò‰Ωç„ÅÆÂá¶ÁêÜ
    for sentence in word_list:
        temp = []
        # Ë™ûÂçò‰Ωç„ÅÆÂá¶ÁêÜ
        for word in sentence:
            word_score = []
            score = dic.get(word)
            word_score = (word, score)
            temp.append(word_score)
        result.append(temp)
    # ÊñáÊØé„Å´„Éá„Éº„Çø„Éï„É¨„Éº„É†Âåñ„Åó„Å¶Ë°®Á§∫
    # for i in range(len(result)):
    #     print(lines[i], '\n', pd.DataFrame(result[i], columns=["word", "score"]), '\n')
    # ÊñáÂçò‰Ωç„ÅÆÂπ≥ÂùáÂÄ§„ÇíË®àÁÆó
    mean_list = []
    for i in result:
        temp = []
        for j in i:
            if not j[1] == None:
                temp.append(float(j[1]))
        mean = (sum(temp) / (len(temp) + 0.00001))
        mean_list_main.append(mean)
  except TypeError:
    mean_list = []
    mean = 0
    mean_list_main.append(mean)
  # „Éá„Éº„Çø„Éï„É¨„Éº„É†Âåñ„Åó„Å¶Ë°®Á§∫
#print(pd.DataFrame(mean_list_main, columns=["mean"]))
```
